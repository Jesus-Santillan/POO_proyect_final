#Diccionario de Trabajo

Segun los puntos se tiene que tomar encuenta lo siguiente:

1.- Se necesita observar y pedir los datos de interes para contextualizarse, tomarse un periodo de 1 a 2 dias (preferentemente).

2.- Recopilar toda la informacion necesaria en relacion a los tipos de analisis que se realizaran (dependiendo de los objetivos deseados) y los datos que se podran recabar o se recabán.

3.- Se analisara el software existente del Saas y con la demas informacion se contemplaran las partes que ya muestran informacion, ademas proponer nuevas.

4.- Organizar esto para ver la foma de crear el modulo de analitica con forme a el Saas y con forme a las posibilidades de las herramientas seleccionadas (tienen que estar adaptadas a lo que se necesita y lo visto en los putos anteriores).

5.- Codificar un software con todo esto recabado, se necesita implementar las formas de recoleccion de datos y pensado en el tratamiento posterior de estos para la analitica.

6.- Mantener la optimizacion y manejo continuo dependiendo de lo que se concidere (con forme al tiempo y recursos con los que se contará).
